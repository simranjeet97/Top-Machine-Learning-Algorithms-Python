## 1. **Linear Regression**
It is the simplest algorithms of machine learning, as its main focus is to determine the relationship between the Independent Variables(one or more then one) and Target Variable (Dependent Variable).

When there is only One Independent Variable it is called **Uni-Variate Linear Regression** and if there are multiple Independent Variables it is called **Multiple Linear Regression.**


<img src="https://miro.medium.com/max/1280/1*LEmBCYAttxS6uI6rEyPLMQ.png" alt ="Linear Regression" width="500"/>

**Mathematical Foundation** -

Let's say the Target Variable is 'Y' and set of all the Independent Variables are 'X = (X1,X2,....Xn)', Here 'n' is the number of Predictors.
So the Equation of Linear Relationship between X and Y is -
    
    Y = B0 + B1X1 + B2X2 + .... + BnXn + Error

'B' is the Regression Coefficients.

So, Here Linera Regression estimates the Regression Coefficients (B0,B1..Bn). The Function should be Written as F(X) = B0 + B1X1 + B2X2 + .... + BnXn, that determines the Dependencies between the input and output variables.

So, Now the Predicted Response  F(Xk), for each observation k = 1....n is very similar to the Actual Value of Yk.

The Difference between the Yk - F(Xk) for all the Observations called Error or Residual.

Regression Analysis is basically all about predicting Weights, that has smallest Residuals.

We use Method of Ordinary Least Squares for best Weights, by minimizing the sum of Sqaured Residuals for all the Observations k = 1....n.
   
    OLS =  Œ£·µ¢(ùë¶·µ¢ - ùëì(ùê±·µ¢))¬≤

The variation of actual responses Yk, k = 1‚Ä¶ùëõ, occurs partly due to the dependence on the predictors Xk.

**Coefficient of Determination (ùëÖ¬≤)** - It states the amount of variation in 'Y' can be explained by the dependence on 'X' using the particular regression model. 

Larger ùëÖ¬≤ indicates a better fit and means that the model can better explain the variation of the output with different inputs.

The value ùëÖ¬≤ = 1 corresponds to OLS = 0, that is to the perfect fit since the values of predicted and actual responses fit completely to each other.

So our main goal is to minimize these Error to make the red squares and green cirles close enough so that our line is perfect for prediction.


**NOTE** - The p-values help determine whether the relationships that you observe in your sample also exist in the larger population. The p-value for each independent variable tests the null hypothesis that the variable has no correlation with the dependent variable.

    A p-value less than 0.05 (typically ‚â§ 0.05) is statistically significant.It indicates strong evidence against the null hypothesis,than a 5% probability the null is correct (and the results are random). Therefore, we reject the null hypothesis, and accept the alternative hypothesis.
    A p-value higher than 0.05 (> 0.05) is not statistically significant and indicates strong evidence for the null hypothesis.

     


```python
#From Scratch Implementation 

import numpy as np
import matplotlib.pyplot as plt
 
def estimate_coef(x, y):
    # number of observations/points
    n = np.size(x)
 
    # mean of x and y vector
    m_x = np.mean(x)
    m_y = np.mean(y)
 
    # calculating cross-deviation and deviation about x
    SS_xy = np.sum(y*x) - n*m_y*m_x
    SS_xx = np.sum(x*x) - n*m_x*m_x
 
    # calculating regression coefficients
    b_1 = SS_xy / SS_xx
    b_0 = m_y - b_1*m_x
 
    return (b_0, b_1)
 
def plot_regression_line(x, y, b):
    # plotting the actual points as scatter plot
    plt.scatter(x, y, color = "m",
               marker = "o", s = 30)
 
    # predicted response vector
    y_pred = b[0] + b[1]*x
 
    # plotting the regression line
    plt.plot(x, y_pred, color = "g")
    plt.scatter(x, y_pred,color = "r")
 
    # putting labels
    plt.xlabel('x')
    plt.ylabel('y')
 
    # function to show plot
    plt.show()
 
def main():
    # observations / data
    x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
    y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
 
    # estimating coefficients
    b = estimate_coef(x, y)
    print("Estimated coefficients:\nb_0 = {}  \
          \nb_1 = {}".format(b[0], b[1]))
 
    # plotting regression line
    plot_regression_line(x, y, b)
 
if __name__ == "__main__":
    main()
```

```python
#Library Implementations 

import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

print("Predicted Values -", mymodel)
print("Coefficient of Determination -",r)
print("P-Value -",p)
print("Error -",std_err)
print("Slope of Line -",slope)
print("Intercept -",intercept)

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
#Prediction 
pred = myfunc(12)
print(pred)
```

###**UNDERFITTING AND OVERFITTING** -

This occurs when we are confuse to decide the optimal degree of the polynomial regression.

**Underfitting** occurs when a model can‚Äôt accurately capture the dependencies among data, usually as a consequence of its own simplicity. It often yields a low ùëÖ¬≤ with known data and bad generalization capabilities when applied with new data.

**Overfitting** happens when a model learns both dependencies among data and random fluctuations. In other words, a model learns the existing data too well. Complex models, which have many features or terms, are often prone to overfitting. When applied to known data, such models usually yield high ùëÖ¬≤. However, they often don‚Äôt generalize well and have significantly lower ùëÖ¬≤ when used with new data.

## **Decision Tree Algorithms**
Decision Tree Algorithms are Supervised Learning Algorithms that are used for Classification and Regression Cases. It works for both Categorical and Continous Data.

Tree based algorithms empower predictive models with high accuracy, stability and ease of interpretation. Unlike linear models, they map non-linear relationships quite well.

In this technique, we split the population or sample into two or more homogeneous sets (or sub-populations) based on most significant splitter / differentiator in input variables.

<img src="https://d1jnx9ba8s6j9r.cloudfront.net/blog/wp-content/uploads/2018/12/Decision-Tree-Machine-Learning-Algorithms-Edureka.png" alt ="Linear Regression" width="800"/>

### **Working**
1. Select the best attribute using Attribute Selection Measures(ASM) to split the records.
2. Make that attribute a decision node and breaks the dataset into smaller subsets.
3. Starts tree building by repeating this process recursively for each child until one of the condition will match:
  1. All the tuples belong to the same attribute value.
  2. There are no more remaining attributes.
  3. There are no more instances.

### **Types of Decision Trees** -

On the basis of the type of target variable we have. It can be of two types:

**1. Categorical Variable Decision Tree**: Decision Tree which has categorical target variable (Yes or No, True or False) then it called as categorical variable decision tree. 

**Example:**- In above scenario of We play or not problem, where the target variables is "PLAY" or "DON'T PLAY".

**2. Continuous Variable Decision Tree**: Decision Tree has continuous target variable( Numbers ) then it is called as Continuous Variable Decision Tree.

**Example:**- Let‚Äôs say we have a problem to predict whether a customer will pay his renewal premium with an insurance company (yes/ no). Here we know that income of customer is a significant variable but insurance company does not have income details for all customers. Now, as we know this is an important variable, then we can build a decision tree to predict customer income based on occupation, product and various other variables. In this case, we are predicting values for continuous variable.

### **Terminologies in Decision Tree Algorithhms**

**1. Root Node:** It represents entire population or sample and this further gets divided into two or more homogeneous sets.

**2. Splitting:** It is a process of dividing a node into two or more sub-nodes.

**3. Decision Node:** When a sub-node splits into further sub-nodes, then it is called decision node.

**4. Leaf/ Terminal Node:** Nodes do not split is called Leaf or Terminal node.

**5. Pruning:** When we remove sub-nodes of a decision node, this process is called pruning. You can say opposite process of splitting.

**6. Branch / Sub-Tree:** A sub section of entire tree is called branch or sub-tree.

**7. Parent and Child Node:** A node, which is divided into sub-nodes is called parent node of sub-nodes where as sub-nodes are the child of parent node.

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
# load dataset
df = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/diabetes.csv")
```

```python
#Feature Selection
#split dataset in features and target variable
feature_cols = ['Pregnancies', 'Insulin', 'BMI', 'Age','Glucose','BloodPressure','DiabetesPedigreeFunction']
X = df[feature_cols] # Features
y = df.Outcome # Target variable
```

It is one of the most simple, straightforward and versatile classification algorithms which is used to solve classification problems.

In statistics, the Logistic Regression model is a widely used statistical model which is primarily used for classification purposes. It means that given a set of observations, Logistic Regression algorithm helps us to classify these observations into two or more discrete classes. So, the target variable is discrete in nature.

The Logistic Regression algorithm works as follows -

Logistic Regression algorithm works by implementing a linear equation with independent or explanatory variables to predict a response value. For example, we consider the example of number of hours studied and probability of passing the exam. Here, number of hours studied is the explanatory variable and it is denoted by x1. Probability of passing the exam is the response or target variable and it is denoted by z.

If we have one explanatory variable (x1) and one response variable (z), then the linear equation would be given mathematically with the following equation-

    z = Œ≤0 + Œ≤1x1  

Here, the coefficients Œ≤0 and Œ≤1 are the parameters of the model.

If there are multiple explanatory variables, then the above equation can be extended to -

    z = Œ≤0 + Œ≤1x1+ Œ≤2x2+‚Ä¶‚Ä¶..+ Œ≤nxn

Here, the coefficients Œ≤0, Œ≤1, Œ≤2 and Œ≤n are the parameters of the model.

So, the predicted response value is given by the above equations and is denoted by z.

## **Sigmoid Function**
This predicted response value, denoted by z is then converted into a probability value that lie between 0 and 1. We use the sigmoid function in order to map predicted values to probability values. This sigmoid function then maps any real value into a probability value between 0 and 1.

In machine learning, sigmoid function is used to map predictions to probabilities. The sigmoid function has an S shaped curve. It is also called sigmoid curve.

## **Decision boundary**
The sigmoid function returns a probability value between 0 and 1. This probability value is then mapped to a discrete class which is either ‚Äú0‚Äù or ‚Äú1‚Äù. In order to map this probability value to a discrete class (pass/fail, yes/no, true/false), we select a threshold value. This threshold value is called Decision boundary. Above this threshold value, we will map the probability values into class 1 and below which we will map values into class 0.

Mathematically, it can be expressed as follows:-

    p ‚â• 0.5 => class = 1

    p < 0.5 => class = 0

Generally, the decision boundary is set to 0.5. So, if the probability value is 0.8 (> 0.5), we will map this observation to class 1. Similarly, if the probability value is 0.2 (< 0.5), we will map this observation to class 0. This is represented in the graph below-

## **Assumptions of Logistic Regression**

The Logistic Regression model requires several key assumptions. These are as follows:-

1. Logistic Regression model requires the dependent variable to be binary, multinomial or ordinal in nature.

2. It requires the observations to be independent of each other. So, the observations should not come from repeated measurements.

3. Logistic Regression algorithm requires little or no multicollinearity among the independent variables. It means that the independent variables should not be too highly correlated with each other.

4. Logistic Regression model assumes linearity of independent variables and log odds.

5. The success of Logistic Regression model depends on the sample sizes. Typically, it requires a large sample size to achieve the high accuracy.

## **Types of Logistic Regression**

1. **Binary Logistic Regression**
In Binary Logistic Regression, the target variable has two possible categories. The common examples of categories are yes or no, good or bad, true or false, spam or no spam and pass or fail.

2. **Multinomial Logistic Regression**
In Multinomial Logistic Regression, the target variable has three or more categories which are not in any particular order. So, there are three or more nominal categories. The examples include the type of categories of fruits - apple, mango, orange and banana.

3. **Ordinal Logistic Regression**
In Ordinal Logistic Regression, the target variable has three or more ordinal categories. So, there is intrinsic order involved with the categories. For example, the student performance can be categorized as poor, average, good and excellent.

```python
def add_age(cols):
    Age = cols[0]
    initial = cols[1]
    if pd.isnull(Age):
        return int(data[data["Initial"] == initial]["Age"].mean())
    else:
        return Age
```

```python
data['Age'] = data[['Age','Initial']].apply(add_age,axis=1)
test_data['Age'] = test_data[['Age','Initial']].apply(add_age,axis=1)
```

```python
def process_age(df,cut_points,label_names):
    df["Age"] = df["Age"].fillna(-0.5)
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df
```

```python
cut_points = [-1,0,5,12,18,35,60,100]
label_names = ["Missing","Infant","Child","Teenager","Young Adult","Adult","Senior"]
train = process_age(data,cut_points,label_names)
test = process_age(test_data,cut_points,label_names)

pivot = train.pivot_table(index="Age_categories",values='Survived')
pivot.plot.bar()
```

```python
def create_dummies(df,column_name):
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df

for column in ["Pclass","Sex","Age_categories",'Embarked']:
    train = create_dummies(train,column)
    test = create_dummies(test,column)
```

# **Multi-Class Logistic Regression**
You can think of multinomial logistic regression as logistic regression (more specifically, binary logistic regression) on steroids. While the binary logistic regression can predict binary outcomes (eg.- yes or no, spam or not spam, 0 or 1, etc.), the MLR can predict one out of k-possible outcomes, where k can be any arbitrary positive integer.

<img src="https://miro.medium.com/max/770/1*dVsfG6i-Y93prmLgTcxRwA.jpeg">

## **How does MLR work?**
The multinomial regression function is a statistical classification algorithm. What this means is that once we feed the function a set of features, the model performs a series of mathematical operations to normalize the input values into a vector of values that follows a probability distribution.

<img src="https://miro.medium.com/max/770/1*23H-Vwy_p43ai0Qvz9d6Pg.jpeg">

- The input that we give to the model is a feature vector, X, containing features x1, x2, x3‚Ä¶..xn.
- The output we get is a probability vector Y, containing probabilities y1, y2, y3‚Ä¶yk for the k target classes.
- Here, y1 + y2 + y3‚Ä¶ + yk = 1, since the total probability of all the possible events in a system is always 1.

Finally, the outcome with the highest probability will be the predicted outcome for the given feature set.
**Now the question is, how exactly does the MLR function convert feature sets to probability values?** We will try to understand this while working on our project.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

np.random.seed(42)

train = pd.read_csv('train.csv')
#print(train['species'])
print("Total Unique Species - ", train['species'].nunique())
train.head()
```

```python
x_train = train.drop(['id', 'species'], axis=1).values
x_train
```

```python
le = LabelEncoder().fit(train['species'])
y_train = le.transform(train['species'])
```

```python
scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
```

```python
params = {'C':[100, 1000], 'tol': [0.001, 0.0001]}
```

```python
log_reg = LogisticRegression(solver='lbfgs', multi_class='multinomial')
clf = GridSearchCV(log_reg, params, scoring='accuracy', refit='True', n_jobs=-1, cv=5)
clf.fit(x_train, y_train)
```

```python
print("best params: " + str(clf.best_params_))
```

```python
test = pd.read_csv('test.csv')
test_ids = test.pop('id')
x_test = test.values
x_test = scaler.transform(x_test)
test.head()
```

```python
y_test = clf.predict_proba(x_test)
result = pd.DataFrame(y_test, index=test_ids, columns=le.classes_)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from statistics import *
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(log_reg, x_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# report the model performance
print('Mean Accuracy: %.3f (%.3f)' % (mean(n_scores), stdev(n_scores)))
```
